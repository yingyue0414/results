```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context("talk")
```

```python
df_2020 = pd.read_pickle("./2020/outputs/2020.pkl")
df_3030 = pd.read_pickle("./3030/outputs/3030.pkl")
df_4040 = pd.read_pickle("./4040/outputs/4040.pkl")
df_5050 = pd.read_pickle("./5050nt/outputs/5050r15nt.pkl")
```

```python
L = [20, 30, 40]
kappa = 20.0
sigma = 0.0
q = np.linspace(0.08, 1.1, 1000)
def hq2(q, L, kappa = 83.4, sigma = 0.0, kbT = 4.17):
    return kbT / (L ** 2 * (kappa * q ** 4 + sigma * q ** 2))
hq2(q, L[0])
```
