```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
sns.set_context("talk")
```

```python
R = 59.0
a = np.pi/4

x = np.linspace(R * np.sin(a), 2 * R * np.sin(a), 400)
y = 2 * R * np.cos(a) - np.sqrt(R**2 - (x - 2 * R * np.sin(a))**2)
theta = np.linspace(0.0, 2 * np.pi, 400)
```

```python
def convert_to_cart(r, z, theta):
    xc = r * np.cos(theta)
    yc = r * np.sin(theta)
    zc = z
    return xc, yc, zc
```

```python
coords = np.zeros([len(x)*len(theta), 3])
index = 0
for xv, yv in zip(x, y):
    for thetav in theta:
        coords[index] = convert_to_cart(xv, yv, thetav)
        index += 1
coords
```

```python
coordst = coords.transpose()
```

```python
xcap = np.linspace(0, R * np.sin(a), 400)
ycap = np.sqrt(R**2 - xcap**2)
thetacap = np.linspace(0.0, 2 * np.pi, 400)

coordscap = np.zeros([len(xcap)*len(thetacap), 3])
index = 0
for xv, yv in zip(xcap, ycap):
    for thetav in thetacap:
        coordscap[index] = convert_to_cart(xv, yv, thetav)
        index += 1
coordscap
```

```python
coordscapt = coordscap.transpose()
```

```python
# Plot the surface.

#%
fig = plt.figure(figsize=(24, 12))
ax = plt.axes(projection='3d')

#surf_gags = ax.plot_trisurf(gags["x"], gags["y"], gags["z"], cmap="Oranges")
#surf_vertices = ax.plot_trisurf(coordscapt[0], coordscapt[1], coordscapt[2], \
#                                cmap="viridis", alpha = 1.0)
surf_vertices = ax.plot_trisurf(coordst[0], coordst[1], coordst[2], \
                                cmap="viridis", alpha = 1.0)

# https://stackoverflow.com/questions/54994600/pyplot-legend-poly3dcollection-object-has-no-attribute-edgecolors2d
surf_vertices._facecolors2d=surf_vertices._facecolors
surf_vertices._edgecolors2d=surf_vertices._edgecolors

#plt.colorbar(surf_vertices, shrink = 0.65, pad = 0.01, anchor = (-0.0, 0.5), label = "height (nm)")

#ax.set_xlabel('X')
#ax.set_ylabel('Y')
#ax.set_zlabel('Z')


ax.set_xlim3d(-90.0, 90.0)
ax.set_ylim3d(-90.0, 90.0)
ax.set_zlim3d(-90.0, 90.0)
ax.set_axis_off()
ax.legend(prop={'size': 12})
ax.view_init(elev = 15, azim = 45)

#plt.savefig(num_gags + "gags_membrane_15deg_uniform_scale.png", dpi = 900)
```
